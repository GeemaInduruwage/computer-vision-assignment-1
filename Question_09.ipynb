### Image Sharpening

I'll use the Pillow (PIL) library to perform image sharpening. First, I'll load the image, then apply a `SHARPEN` filter, and finally display both the original and the sharpened version.

```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt

# Load the image
image_path = '/content/einstein.png'
original_image = Image.open(image_path)

# Apply the sharpening filter
sharpened_image = original_image.filter(ImageFilter.SHARPEN)

# Display the original and sharpened images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(original_image)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Sharpened Image')
plt.imshow(sharpened_image)
plt.axis('off')

plt.tight_layout()
plt.show()
```
