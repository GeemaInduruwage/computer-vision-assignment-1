# Question 08
8. Fig. 4 is an image corrupted with salt and pepper noise.
(a) Apply Gaussian smoothing.
(b) Apply median filtering.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
img_path = '/content/Picture1.jpg'
img = cv2.imread(img_path)
if img is None:
    print(f'Error: Could not load image at {img_path}')
else:
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # (a) Apply Gaussian Smoothing
    # Using a 5x5 kernel
    gaussian_blur = cv2.GaussianBlur(img_rgb, (5, 5), 0)

    # (b) Apply Median Filtering
    # Median filter is highly effective against salt and pepper noise
    median_blur = cv2.medianBlur(img_rgb, 5)

    # Display the results
    plt.figure(figsize=(18, 6))

    plt.subplot(1, 3, 1)
    plt.imshow(img_rgb)
    plt.title('Original (Salt & Pepper Noise)')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(gaussian_blur)
    plt.title('Gaussian Smoothing')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(median_blur)
    plt.title('Median Filtering')
    plt.axis('off')

    plt.show()
```
